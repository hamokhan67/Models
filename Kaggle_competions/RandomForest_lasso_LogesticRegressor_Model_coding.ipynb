```python
#training and testing spliting
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
y_test = y_test.astype('int64')
y_test
```

```python
pipelines = {
    'rf': Pipeline(steps=[('scaler', StandardScaler()), ('model', RandomForestRegressor())]), 
    'lasso': Pipeline(steps=[('scaler', StandardScaler()), ('model', Lasso())]),
    'logreg': Pipeline(steps=[('scaler', StandardScaler()), ('model', LogisticRegression())])
}
```

```python
param_grids = {
    'rf': {
        'model__n_estimators': [100],
        #'model__max_depth': [None, 10, 20],
        #'model__min_samples_split': [2, 5, 10],
    },
    'lasso': {
        'model__alpha': [0.01, 0.1, 1, 10],
        'model__max_iter': [400,200,500],
        #'model__tol': [0.0001, 0.001, 0.01],
    },
    'logreg': {
        #'model__penalty': ['l1', 'l2', 'elasticnet', 'none'],
        #'model__C': [0.01, 0.1, 1, 10, 100],
        #'model__solver': ['saga'],
        'model__max_iter': [500],
        #'model__l1_ratio': [0.0, 0.5, 1.0],
    }
}
```

```python
fit_models = {}

cv_strategy = StratifiedKFold(n_splits=3)

for algo,pipelin in pipelines.items():
    print(f'Traning the {algo} model.')
    model = GridSearchCV(pipelin, param_grids[algo], cv=3, n_jobs=-1)
    print(model)
    model.fit(X_train, y_train)
    fit_models[algo] = model
for algo, model in fit_models.items():
    print(f'{algo} model best parameters: {model.best_params_}')
    print(f'{algo} model best cross-validation score: {model.best_score_:.2f}')
```

```python
fit_models
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
for algo,model in fit_models.items():
    yhat = model.predict(X_test)
    mabser = mean_absolute_error(y_test,yhat)
    msqer = mean_squared_error(y_test,yhat)
    r2 = r2_score(y_test,yhat)
    print(f'Metrics for {algo}: Mean Absolute Error: {mabser}, Mean Squared Error: {mabser}, R2 Score: {r2} ')
```
